```python
import pandas as pd
df=pd.read_csv("/content/sample_data/carvana.csv")
```

```python
df['Year']=df['Year'].astype(str).str[:4].astype(int)
```

```python
df[df['Year']>2026]
```

```python
df.rename(columns={"Miles":"Kilometers"},inplace=True)
df['Kilometers']=(df['Kilometers']* 1.6).astype(int)
```

```python
from sklearn.preprocessing import StandardScaler as ss
scaler=ss()
df['Kilometers']=scaler.fit_transform(df[['Kilometers']])
```

from here it's linear regression model which gives a huge error

```python
nameDummies=pd.get_dummies(df['Name'])
```

```python
df=pd.concat([df,nameDummies],axis=1)
```

```python
df.drop('Name',axis=1,inplace=True)
```

```python
y=df['Price']
x=df.drop('Price',axis=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
xTrain,xTest,yTrain,yTest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(xTrain,yTrain)
```

```python
from numpy._core.fromnumeric import mean
yPred=model.predict(xTest)
mSe=mean_squared_error(yTest,yPred)
print(int(mSe))
```

```python
from sklearn.preprocessing import LabelEncoder
Le=LabelEncoder()
df['encodedName']=Le.fit_transform(df['Name'])
df.drop('Name',axis=1,inplace=True)
```

```python
x=df.drop('Price',axis=1)
y=df['Price']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
xTrain,xTest,yTrain,yTest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=RandomForestRegressor(n_estimators=100,random_state=42)
model.fit(xTrain,yTrain)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
yPred=model.predict(xTest)
mAe=mean_absolute_error(yTest,yPred)
r=r2_score(yTest,yPred)
print("$"+str(mAe))
print(int(r))
```
